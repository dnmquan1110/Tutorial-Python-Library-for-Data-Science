```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
from bs4 import BeautifulSoup
from urllib.request import urlopen
import re
```

```python
url = 'https://hubertiming.com/results/2018MLK'
html=urlopen(url)
```

```python
soup = BeautifulSoup(html)
```

```python
title = soup.title
print(title)
print(title.text)
```

```python
links = soup.find_all('a',href=True)
for link in links:
    print(link.get('href'))
```

```python
allrows = soup.find_all('tr')
data = []
for row in allrows:
    row_list = row.find_all('td')
    dataRow = []
    for cell in row_list:
        dataRow.append(cell.text)
    data.append(dataRow)

titles = data[0]
data = data[1:]
print(data[-2:])
```

```python
df = pd.DataFrame(data)
df.head(10)
```

```python
col_headers = soup.find_all('th')
header_list = []
for col in col_headers:
    header_list.append(col.text)
header_list
```

```python
df.columns = header_list
df.head()
```
